```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select


servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
import pandas as pd

notas_dataframe = pd.read_excel('NotasEmitir.xlsx')
notas_dataframe.head()
```

```python
for linha in notas_dataframe.index:
    navegador.find_element(By.XPATH, '//*[@id="nome"]').send_keys(notas_dataframe.loc[linha, 'Cliente'])
    navegador.find_element(By.NAME, 'endereco').send_keys(notas_dataframe.loc[linha, 'Endereço'])
    navegador.find_element(By.NAME, 'bairro').send_keys(notas_dataframe.loc[linha, 'Bairro'])
    navegador.find_element(By.NAME, 'municipio').send_keys(notas_dataframe.loc[linha, 'Municipio'])
    navegador.find_element(By.NAME, 'cep').send_keys(str(notas_dataframe.loc[linha, 'CEP']))
    elemento = navegador.find_element(By.TAG_NAME, 'select')
    elemento_select = Select(elemento)
    elemento_select.select_by_visible_text(notas_dataframe.loc[linha, 'UF'])
    navegador.find_element(By.NAME, 'cnpj').send_keys(str(notas_dataframe.loc[linha, 'CPF/CNPJ']))
    navegador.find_element(By.NAME, 'inscricao').send_keys(str(notas_dataframe.loc[linha, 'Inscricao Estadual']))
    navegador.find_element(By.NAME, 'descricao').send_keys(notas_dataframe.loc[linha, 'Descrição'])
    navegador.find_element(By.NAME, 'quantidade').send_keys(str(notas_dataframe.loc[linha, 'Quantidade']))
    navegador.find_element(By.NAME, 'valor_unitario').send_keys(str(notas_dataframe.loc[linha, 'Valor Unitario']))
    navegador.find_element(By.NAME, 'total').send_keys(str(notas_dataframe.loc[linha, 'Valor Total']))
    navegador.find_element(By.XPATH, '//*[@id="formulario"]/button').click()
    navegador.refresh()



```
